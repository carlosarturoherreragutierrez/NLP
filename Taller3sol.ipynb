# Punto 1:

- `[15 pts]` Hacer Web Scraping de 10 animales en Wikipedia (en búcle)
- `[10 pts]` Obtener el **encabezado** de cada animal
- `[15 pts]` Obtener todos los **textos** que están en las etiquetas de negrilla.

```python
import bs4 as bs 
import urllib.request
import re
```

```python
animal=['Felis_silvestris_catus','Equus_ferus_caballus', 'Spheniscidae', 'Canis_familiaris','Megaptera_novaeangliae','Gallus_gallus_domesticus','Hydrochoerus_hydrochaeris','Eudocimus_ruber','Canguro','Equus_africanus_asinus']
```

```python
paginas_html=[]

#Procesar la página web como HTML y muestra encabezados de páginas
def visita_paginas(indice,paginas):
 
        pagina = 'https://es.wikipedia.org/wiki/'+paginas[indice]
        request = urllib.request.Request(pagina)
        webpage = urllib.request.urlopen(request)
        source = webpage.read()
        webpage.close()      
        soup = bs.BeautifulSoup(source, 'html.parser')
        categoria= soup.find("title").text
        paginas_html.append(soup)
        patron ="-"
        pos=re.search(patron, categoria).span()
        print(f"{indice+1}.Encabezado: {categoria[0:pos[0]-1]}")
        
    
#Listado de palabras en negrita de cada animal
def palabras_negrita_especie(especie):
    #print("palabras en negrilla:")
    for l in especie.find("p").find_all("b"):
        print(l.text)

#Reemplazar tildes con asteriscos
def reemplazo_tildes(especie):
    for l in especie.find("p").find_all("b"):
        palabra=l.text
        cambio=re.sub(r"[áéíóúÁÉÍÓÚ]","*",palabra)
        print(cambio)
 
    
#Listado de encabezados de página y palabras negritas 
for k in range(0,10):
    visita_paginas(k,animal)
    palabras_negrita_especie(paginas_html[k])
    print(" ")
```

# Punto 2:
- `[10 pts]` Usando regex, reemplazar todas las tíldes del punto anterior por un asterisco (¡Ojo, los espacios se quedan!)


```python
for k in range(0,10):
    reemplazo_tildes(paginas_html[k])
    print(" ")
```
